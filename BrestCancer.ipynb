```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset from sklearn
brest_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(brest_cancer_dataset)
```

```python
# loading the dta to dataframe
data_frame = pd.DataFrame(brest_cancer_dataset.data, columns = brest_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = brest_cancer_dataset.target
```

```python
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# statistical measures
data_frame.describe()
```

```python
# checking this distribution of target variable
data_frame['label'].value_counts()
```

1 -> Benign
0 -> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and target

```python
x = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
print(x)
```

Splitting the data into training data and test

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Standardize the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
print(x_train_std)
```

Building Neural Network

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# setting up the neural network
history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=10)
```

Visualizing Accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper left')
plt.show()
```

Accuracy of the model on the test data

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print('Test accuracy:', accuracy)
```

**Building the predictive system**

```python
input_data = (11.42,20.38,77.58,386.1,0.1425,0.2839,0.2414,0.1052,0.2597,0.09744,0.4956,1.156,3.445,27.23,0.00911,0.07458,0.05661,0.01867,0.05963,0.009208,14.91,26.5,98.87,567.7,0.2098,0.8663,0.6869,0.2575,0.6638,0.173
)

# change the input data to numpy array
input_data_as_numpy_array = np.array(input_data)

# reshape the numpy array as we are predicting for one label point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# std the input data
input_data_std = scaler.transform(input_data_reshaped)

# prediction
prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if (prediction_label[0] == 0):
  print('The breast cancer is Malignant')
else:
  print('The breast cancer is Benign')
```
